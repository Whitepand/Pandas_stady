```python
import pandas as pd
import numpy as np
#rating_movie_data = pd.read_csv('D:/Python обучение/Pandas_learning/Pandas/Pandas_stady/Pandas_v5/movies_data/ratings_movies.csv', sep=',')
rating_movie_data = pd.read_csv('C:/Users/LeonidKogtev/Python_learning/Pandas/practice/Pandas_stady/Pandas_v5/movies_data/ratings_movies.csv', sep=',')
display(rating_movie_data.head())
display(rating_movie_data.tail())
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
rating_movie_data['year_release'] = rating_movie_data['title'].apply(get_year_release)
rating_movie_data['year_release'] = rating_movie_data['year_release'].astype('Int64')
mask_1999 = rating_movie_data['year_release'] == 1999
rating_movie_data_1999 = rating_movie_data[mask_1999]
mean_reting_1999 = rating_movie_data_1999.groupby(by='title')['rating'].mean()
#display(rating_movie_data_1999.groupby(by='title')['rating'].mean())
display(mean_reting_1999.sort_values())
#display(rating_movie_data)
#display(rating_movie_data_1999)
```
