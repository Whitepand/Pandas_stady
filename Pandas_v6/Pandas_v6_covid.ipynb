```python
import pandas as pd
covid_data = pd.read_csv('D:/Python обучение/Pandas_learning/Pandas/Pandas_stady/data/covid_data/covid_data.csv')
display(covid_data.head())
```

```python
vaccinations_data = pd.read_csv('D:/Python обучение/Pandas_learning/Pandas/Pandas_stady/data/covid_data/country_vaccinations.csv')
vaccinations_data = vaccinations_data[
    ['country', 'date', 'total_vaccinations', 
     'people_vaccinated', 'people_vaccinated_per_hundred',
     'people_fully_vaccinated', 'people_fully_vaccinated_per_hundred',
     'daily_vaccinations', 'vaccines']
]
#display(vaccinations_data)
```

```python
covid_data = covid_data.groupby(
    ['date', 'country'], 
    as_index=False
)[['confirmed', 'deaths', 'recovered']].sum()


covid_data['date'] = pd.to_datetime(covid_data['date'])
```

```python
vaccinations_data["active"] = covid_data['confirmed'] - covid_data['deaths'] - covid_data['recovered']
```

```python
covid_data = covid_data.sort_values(by=['country', 'date'])
covid_data['daily_confirmed'] = covid_data.groupby('country')['confirmed'].diff()
covid_data['daily_deaths'] = covid_data.groupby('country')['deaths'].diff()
covid_data['daily_recovered'] = covid_data.groupby('country')['recovered'].diff()
```

```python
vaccinations_data['date'] = pd.to_datetime(vaccinations_data['date'])
```

```python
display(covid_data['date'].sort_values())
display(vaccinations_data['date'].sort_values())
display(vaccinations_data['date'].max())
```

```python
display(covid_data)
display(vaccinations_data)
```

```python
covid_df = covid_data.merge(
    vaccinations_data,
    how='left'
)
display(covid_df)
```

```python
#covid_df['death_rate'] =  covid_df['deaths'].sum() / covid_df['confirmed'].sum()  * 100
us_covid = covid_df[covid_df['country'] == 'United States']
rus_covid = covid_df[covid_df['country'] == 'Russia']
us_covid['death_rate'] = us_covid['deaths'] / us_covid['confirmed'] * 100 
display(us_covid['death_rate'].max())
covid_df['recover_rate'] = covid_df['recovered'] / covid_df['confirmed']  * 100
display(rus_covid['recover_rate'].mean())
```
