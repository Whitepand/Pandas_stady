```python
import pandas as pd
import numpy as np
train_data = pd.read_csv('data/citibike-tripdata.csv')
train_df = train_data.copy()
display(train_df.head())


```

```python
display(train_df['usertype'].value_counts(normalize=True))
display(train_df['start station name'].value_counts(normalize=True))
```

```python
train_df_1 = train_df[(train_df["gender"] == 1)]
train_df_2 = train_df[train_df["gender"] == 2]
display(train_df_1.shape[0])
display(train_df_2.shape[0])
display(train_df['gender'].mode())
```

```python
drop_id = ["start station id", "end station id"]
train_df_drop2 = train_df.drop(drop_id, axis = 1)
display(train_df_drop2)
```

```python
train_df['Age'] = 2018 - train_df['birth year']
train_df_drop2 = train_df.drop(['birth year'], axis = 1)
display(train_df_drop2[(train_df_drop2["Age"] > 60)].shape[0])
display(train_df_drop2.head())
```

```python
train_df_drop2['starttime'] = pd.to_datetime(train_df_drop2['starttime'])
train_df_drop2['stoptime'] = pd.to_datetime(train_df_drop2['stoptime'])

train_df_drop2_maines = train_df_drop2['stoptime'] - train_df_drop2['starttime']
train_df_drop2["trip duration"] = train_df_drop2_maines 
display(train_df_drop2.iloc[3])
```

```python
train_df_drop2['weekend'] = train_df_drop2['starttime'].dt.dayofweek
display(train_df_drop2.head())
train_df_drop2_masl_weekend = train_df_drop2[train_df_drop2['weekend'] >= 5]
#display(train_df_drop2[train_df_drop2['weekend'] >= 5].shape[0])

display(train_df_drop2.head())
```

```python
def get_weekday(weekday):
    if weekday >= 5:
        return 1
    else:
        return 0
train_df_drop2['weekend'] =  train_df_drop2['weekend'].apply(get_weekday) 
display(train_df_drop2)
```

```python
train_df_mask_1 = train_df_drop2[(train_df_drop2['weekend'] == 1)]
train_df_mask_0 = train_df_drop2[(train_df_drop2['weekend'] == 0)]
display(train_df_mask_1.shape[0])
display(train_df_mask_0.shape[0])
```

```python
def get_time_of_day(time):
    if time >= 0 and time <= 6:
        return 'night'
    elif time > 6 and time <= 12:
        return 'morning'
    elif time > 12 and time <= 18:
        return 'day'
    elif time > 18 and time <= 23:
        return 'evening'

train_df_drop2['time_of_day'] =  train_df_drop2['starttime'].dt.hour.apply(get_time_of_day)   
```

```python
train_df_drop2_mask_day = train_df_drop2[(train_df_drop2['time_of_day'] == 'day')]
train_df_drop2_mask_night = train_df_drop2[(train_df_drop2['time_of_day'] == 'night')]
#display(train_df_drop2_mask_night)


train = train_df_drop2_mask_day.shape[0] / train_df_drop2_mask_night.shape[0]
print(train)
```
